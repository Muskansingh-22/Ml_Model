```python
# Data Manipulation
import pandas as pd
import numpy as np
# Data Visualisation
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import logging
logging.basicConfig(level = logging.INFO,
                    filemode= 'w',
                    filename= 'model.log',
                    format= '%(asctime)s - %(levelname)s - %(message)s',force= True)
```

```python
url = 'https://raw.githubusercontent.com/Muskansingh-22/Ml_Model/refs/heads/main/Concrete_Data.csv'

df = pd.read_csv(url)

df
```

```python
x = df.drop(columns = 'Concrete compressive strength(MPa, megapascals) ',axis = 1)
y = df['Concrete compressive strength(MPa, megapascals) ']
from sklearn.preprocessing import StandardScaler,MinMaxScaler,RobustScaler
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state= 42)
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)  # Seen data
x_test = scaler.transform(x_test)        # Unseen data
```

```python
# Using linear Regression

from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(x_train,y_train)
y_pred_LR = LR.predict(x_test)
from sklearn.metrics import r2_score
r2_score_LR = r2_score(y_test,y_pred_LR)
r2_score_LR
```

```python

from sklearn.ensemble import RandomForestRegressor
RF = RandomForestRegressor()
RF.fit(x_train,y_train)
y_pred_RF = RF.predict(x_test)
from sklearn.metrics import r2_score
r2_score_RF= r2_score(y_test,y_pred_RF)
r2_score_RF
```

```python
# Saving model as Pickle

import pickle
filename = 'concrete_strength_model.pkl'
pickle.dump(RF,open(filename,'wb'))
```
